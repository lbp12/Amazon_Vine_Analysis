```python
# import dependencies
import pandas as pd
```

```python
vine_df = pd.read_csv("vine_table.csv")
vine_df.head()
```

```python
vine_df.count()
```

```python
# Create a DataFrame or table where there are 20 or more total votes
filtered_total_votes = vine_df.loc[vine_df['total_votes'] >= 20]
filtered_total_votes.head()
```

```python
# Create a DataFrame or table where the percentage of helpful_votes is equal to or greater than 50%
helpful_votes = filtered_total_votes.loc[(filtered_total_votes['helpful_votes']/filtered_total_votes['total_votes']) >= 0.5]
helpful_votes.head()
```

```python
#  Create a DataFrame or table where there is a Vine review
paid_review = helpful_votes.loc[helpful_votes['vine'] == 'Y']
paid_review.head()
```

```python
# Create a DataFrame or table where there isnâ€™t a Vine review
unpaid_review = helpful_votes.loc[helpful_votes['vine'] == 'N']
unpaid_review.head()
```

```python
# The total number of reviews
paid_count = paid_review["star_rating"].count()



# The number of 5-star reviews
paid_review_5_star = paid_review.loc[paid_review['star_rating'] == 5]['star_rating'].count()

# Percentage of 5-star reviews

paid_review_percent = round(((paid_review_5_star/paid_count) * 100),2)

print(paid_count)
print(paid_review_5_star)
print(paid_review_percent)
```

```python
# The total number of reviews
unpaid_count = unpaid_review["star_rating"].count()



# The number of 5-star reviews
unpaid_review_5_star = unpaid_review.loc[unpaid_review['star_rating'] == 5]['star_rating'].count()

# Percentage of 5-star reviews

unpaid_review_percent = round(((unpaid_review_5_star/unpaid_count) * 100),2)

print(unpaid_count)
print(unpaid_review_5_star)
print(unpaid_review_percent)
```
